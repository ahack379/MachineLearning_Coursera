## K-MEANS CLUSTERING

```python
a = sio.loadmat('ex7data2.mat')

X = a['X'] 
print "%i examples, %i dimensions " % (X.shape[0],X.shape[1])
```

```python
def findClosestCentroids(X,centroids):
    
    idx = []
    
    for x in X:
        min_dist = 1e9
        min_c_it = -1
        c_it = 0
        
        for c in centroids:
            #sq_dist = math.pow(x[0] - c[0],2) + math.pow(x[1] - c[1],2)
            sq_dist = np.sum( np.square(x - c) )
            
            if sq_dist < min_dist:
                min_dist = sq_dist
                min_c_it = c_it
            c_it += 1
        
        idx.append(min_c_it)
    
    return idx

def computeCentroids(X,centroids,idx_v):
    
    new_cents = np.array([])
    
    for c in xrange(centroids.shape[0]):
   
        x_per_c = [ X[i] for i in xrange(len(idx_v)) if idx_v[i] == c ]
        
        sum_c = np.cumsum(x_per_c,axis=0)[-1] / len(x_per_c)
        new_cents = np.append(new_cents,sum_c)
    
    new_cents = new_cents.reshape(centroids.shape[0],X.shape[1])
    
    return new_cents
```

```python
centroids = np.array([[3,3], [6,2], [8,5]])
idx_v = findClosestCentroids(X,centroids)

print "First 3 data points are closest to clusters ", idx_v[:3]
```

```python
#Plot general data
def plotData(X):
    
    plt.figure(figsize=(5,5))
    plt.plot(X[:,0],X[:,1],'ko', mfc='none',markersize=3)
    plt.grid(True)
    plt.title("Data Display Before KMeans")
    plt.show()
    
#Plot shows 3 distinct clusters
#Come back to try: OpenCV dilation, blur, threshold, contour finding
#plotData(X)

# Plot the kmeans outcome
def plotKMeansData(X,centroids,idx_v,mpl_color=False):
    
    plt.figure(figsize=(6,6))
    colors = ['r','g','b']

    for c in xrange(centroids.shape[0]):
        c_it = np.array([ X[i] for i in xrange(len(idx_v)) if idx_v[i] == c ])
        if mpl_color:
            plt.plot(c_it[:,0],c_it[:,1],'o',ms=5,label="%i Clus"%c,alpha = 0.35)
        else:
            plt.plot(c_it[:,0],c_it[:,1],'o',ms=5,label="%i Clus"%c,mec=colors[c],mfc='None')
        plt.grid(True)
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05))
```

```python
def runkMeans(X,iterations,k):

    k_init = random.sample(range(0,X.shape[0]),k)
    centroids = np.array([X[i] for i in k_init])

    prev_centroids = centroids

    for i in xrange(iterations):

        idx_v = findClosestCentroids(X,centroids)
        centroids = computeCentroids(X,centroids,idx_v)  
        prev_centroids = np.append(prev_centroids,centroids,axis=0)
        
    return centroids, idx_v, prev_centroids

iterations, k = 7, 3
centroids, idx_v, prev_centroids = runkMeans(X,iterations,k)

plotKMeansData(X,centroids,idx_v)

c0_v = prev_centroids[0::3]
c1_v = prev_centroids[1::3]
c2_v = prev_centroids[2::3]

plt.plot(c0_v[:,0],c0_v[:,1],'kx-')
plt.plot(c1_v[:,0],c1_v[:,1],'kx-')
plt.plot(c2_v[:,0],c2_v[:,1],'kx-')
plt.title("KMeans Clustering, %i Iterations"%iterations)
```

## 2 PCA STUFF 


```python
pca = sio.loadmat('ex7data1.mat')

X = pca['X'] 
print "%i examples, %i dimensions " % (X.shape[0],X.shape[1])
#plt.figure(figsize=(5,5))
#plt.plot(X[:,0],X[:,1],'o',ms=5,mec='b',mfc='None')
#plt.title("Example Dataset 1")
#plt.show()
```

```python
def featureNormalize(X):
    mean = np.mean(X,axis=0)
    std  = np.std(X,axis=0)
    X_norm = (X - mean) /std
    
    return X_norm, mean, std

def computeCovariance(X):
    m = X.shape[0]
    return np.dot(X.T,X) / m
```

```python
#https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.svd.html
X_norm, mean, std = featureNormalize(X)

cov = computeCovariance(X_norm)
U, s, V = np.linalg.svd(cov, full_matrices=True)

print "Principal Componets are ", U[:,0]
```

```python
x_princ = [mean[0],mean[0] + 1.5 * s[0] * U[0][0]]
y_princ = [mean[1],mean[1] + 1.5 * s[0] * U[0][1]]

x_sec = [mean[0],mean[0] + 1.5 * s[1] * U[1][0]]
y_sec = [mean[1],mean[1] + 1.5 * s[1] * U[1][1]]


plt.figure(figsize=(5,5))
# I can't remember where the 1.5 comes from, but it's in ex7_pca.m
plt.plot(X[:,0], X[:,1],'o',ms=5,mec='b',mfc='None')
plt.plot(x_princ,y_princ,'orange',label='Principal Component',lw=3)
plt.plot(x_sec,y_sec,'g-',label='Secondary Component',lw=3)
plt.ylim(2,8)
plt.xlim(0,7)
plt.legend(loc=0)
plt.show()
```

```python
def projectData(X_norm,U,K):

    return np.dot(X_norm,U[:,:K])

def recoverData(X_apprx,U,K):
    return np.dot(X_apprx,U[:,:K].T)

proj = projectData(X_norm,U,1)
print "First projected entry should be close to 1.481: %.3f" % float(proj[0])

reco = recoverData(proj,U,1)
print "First recovered entry should be close to [-1.047 -1.047] :\
      [%.3f, %.3f]" % (reco[0][0],reco[0][1])

plt.figure()
plt.plot(X_norm[:,0], X_norm[:,1],'o',ms=5,mec='b',mfc='None',label='Original Data')
plt.plot(reco[:,0],reco[:,1],'o',ms=5,mec='r',mfc='None',label='Reduced Data')
plt.ylim(-4,3)
plt.xlim(-4,3)
plt.legend(loc=0)
plt.show()
```

```python
def displayData(data,rows,dim):

    data_32x32 = [ np.reshape(data[im],(32,32),order='F') for im in xrange(rows) ] 
    images = [ smp.toimage( data_32x32[i]) for i in xrange(len(data_32x32)) ]

    plt.figure(figsize=(7,7))
    for i in xrange(len(images)):
        plt.subplot(dim,dim,i+1)
        plt.imshow(images[i],cmap='Greys_r')
        plt.axis('off')
    
    plt.show()

displayData(np.array(faces),100,10)
```

```python
f_norm,_,_ = featureNormalize(faces)

f_cov = computeCovariance(f_norm)
U, s, V = np.linalg.svd(f_cov, full_matrices=True)
```

```python
pics = 36
displayData(U[:,:36].T,pics,math.sqrt(pics))
```

```python
# With K=1, hones in on black + white.  Pretty dope
f_proj = projectData(f_norm[:100,:],U[:,:100],1)
f_reco = recoverData(f_proj,U[:,:100],1)

displayData(f_reco,25,5)

# With K=36, more definition
f_proj_36 = projectData(f_norm[:100,:],U[:,:100],36)
f_reco_36 = recoverData(f_proj_36,U[:,:100],36)

displayData(f_reco_36,25,5)
```
